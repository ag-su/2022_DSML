# 과제 3

```python
date_list = []

for cd, date in zip(df2.index, df2.date):
    start_date = '20171201' 
    end_date = '20210130' 
    stock = fdr.DataReader(cd, start = start_date, end = end_date)
    stock.reset_index(inplace=True)
    
    D9_index = stock[stock['Date'] == str(date)].index - 9 
    D9_date = stock.iloc[D9_index]['Date']
    next_index = stock[stock['Date'] == str(date)].index + 1 
    next_date = stock.iloc[next_index]['Date']
    
    date_list.append([D9_date, next_date])
```

```python
# 최종 머신러닝 데이터의 데이터 프레임 생성을 위한 dictionary 저장
data_dict = {'code': [], 'd0': [], 'info': [], 'up': []}
for i, cd in enumerate(df2.index):
    
    print(i) # 진행 확인
    
    # 종목코드 (code)
    data_dict['code'].append(cd) 
    
    
    # 기준일 (d0)
    data_dict['d0'].append(df2.iloc[i].date)
    
    start_date = date_list[i][0][date_list[i][0].index[0]].date()
    end_date = date_list[i][1][date_list[i][1].index[0]].date()
    stock = fdr.DataReader(cd, start = start_date, end = end_date)
    stock['trading'] = stock['Close'] * stock['Volume']
   
    
    # 10일간의 데이터 (info)
    info_list = []
    for i in range(10):        
        info_list.append(stock.iloc[i, [0, 1, 2, 3, -1]].to_list())
    remove_list=['[', ']']
    for i in range(2): 
        info_list = f'{info_list}'.replace(remove_list[i], '')
    data_dict['info'].append(info_list)    
        
        
    # D+1 종가 2% 상승 여부 (up)
    up = stock.iloc[-2]['Close'] + 0.02 * stock.iloc[-2]['Close']
    
    if stock.iloc[-1]['Close'] >= up: 
        data_dict['up'].append(1)
    else: 
        data_dict['up'].append(0)
```

```python
# 데이터프레임 생성
df3 = pd.DataFrame(data_dict)
df3
```
